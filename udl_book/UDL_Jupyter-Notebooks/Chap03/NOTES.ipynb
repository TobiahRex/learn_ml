```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(x, y1)
ax1.set_title('Sine')

ax2.plot(x, y2)
ax2.set_title('Cosine')

plt.tight_layout()
plt.show()
```
