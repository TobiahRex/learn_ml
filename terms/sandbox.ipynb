```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some log-normally distributed data
mu, sigma = 3.0, 1.0  # mean and standard deviation
s = np.random.lognormal(mu, sigma, 1000)
print(s)

# Apply the logarithm transformation
s_transformed = np.log(s)

# Plot the original and transformed data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.hist(s, bins=50)
ax1.set_title("Original (log-normal)")

ax2.hist(s_transformed, bins=50)
ax2.set_title("Transformed (normal)")

plt.show()
```
